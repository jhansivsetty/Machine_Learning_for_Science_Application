```python
import numpy as np
import random
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#learning rate
alp = 0.1
ep = 1.e-6
```

```python
m = 51
x = np.linspace(0,1,m)
noise = np.random.normal(0, 0.4, x.shape)
y = 15*x + 5 + noise
```

```python
plt.scatter(x,y)
```

```python
w0 = random.random()
w1 = random.random()
```

```python
alp = 1.5
J= [0]
J.append(0.5/m * sum((y-w1*x-w0)**2))
err = 1
iter = 1
```

```python
yh = w1*x +w0
DJ0 = (yh-y)
dw0 = -alp*sum(DJ0)/m
DJ1 = (yh-y)*x
dw1 = -alp*sum(DJ1)/m
w0 = w0 +dw0
w1 = w1+dw1

iter = iter + 1
J.append(0.5/m * sum((y-w1*x-w0)**2))
err = abs(J[iter]-J[iter-1])
```

```python
while err>1.e-5:
    yh = w1*x +w0
    DJ0 = (yh-y)
    dw0 = -alp*sum(DJ0)/m
    DJ1 = (yh-y)*x
    dw1 = -alp*sum(DJ1)/m
    w0 = w0 +dw0
    w1 = w1+dw1
    
    iter = iter + 1
    J.append(0.5/m * sum((y-w1*x-w0)**2))
    err = abs(J[iter]-J[iter-1])
    #err = np.linalg.norm([dw0,dw1])
```
