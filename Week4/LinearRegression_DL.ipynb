```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
x = np.arange(80,-341,-20)
y = [6.47, 6.36, 6.24, 6.12,6.00,5.86, 5.72,5.58,5.43,5.28,5.09,4.91,4.72,4.52,4.3,4.08,3.83,3.58,3.33,3.07,2.76,2.45]
```

```python
plt.scatter(x,y)
```

```python
# linear equation
coeff1 = np.polyfit(x, y, 1)
coeff1
```

```python
# quadratic equation
coeff2 = np.polyfit(x, y, 2)
coeff2
```

```python
# cubic equation
coeff3 = np.polyfit(x, y, 3)
coeff3
```

```python
pred1 = np.polyval(coeff1,x)
pred1
```

```python
plt.scatter(x,y)
plt.plot(x,pred1)
plt.show()
```

```python
pred2 = np.polyval(coeff2,x)
plt.scatter(x,y)
plt.plot(x,pred2)
plt.show()
```

```python
pred3 = np.polyval(coeff3, x)
plt.scatter(x,y)
plt.plot(x,pred3)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x, pred1)
plt.plot(x, pred2, 'r')
plt.plot(x, pred3, 'g')
plt.legend(["Linear fit", "quadratic fit", "cubic fit"])
plt.title("Thermal expansion coefficient")
plt.xlabel("Temperature")
plt.ylabel("Coefficient")
```

Predicting the Termal coefficient

```python
xp = 70
predict1 =  np.polyval(coeff1, xp)
predict2=  np.polyval(coeff2,xp)
predict3=  np.polyval(coeff3,xp)
print(predict1)
print(predict2)
print(predict3)
```
